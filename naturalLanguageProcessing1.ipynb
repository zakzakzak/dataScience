```python
import nltk 
from nltk import sent_tokenize 
from nltk import word_tokenize

nltk.download("popular", quiet=True)
```

```python
# Tokenize the text by sentences 
sentences = sent_tokenize(text)
print(len(sentences))

# print the sentences
sentences
```

```python
# Tokenize the text with words
words = word_tokenize(text)

# how many words are there 
print(len(words))
print("\n")

# print words :
print(words)
```

```python
# import required libraries
from nltk.probability import FreqDist 

# find the frequency : 
fdist = FreqDist(words)

fdist.most_common(10)
```

```python
# Tokenize itu bukan encoding
# yaitu adalah memecah teks menjadi unit kecil (token)
# kalimat : sentence token
# kata : word token
# contoh : text = "I love NLP."
# menjadi : ['I', 'love', 'NLP', '.']
# lalu kenapa tidak hanya sekedar .split(" ") ?
# karena jika case nya "Dr. Smith lives in the U.S."
# maka ketika di pecah menjadi ['Dr', 'Smith lives in the U', 'S'] (berdasarkan titik)
# isi nltk itu adalah kamus kata2 tersebut
```

```python
# plot the graph for fdist  : 
import matplotlib.pyplot as plt

fdist.plot(10)
```

```python
# there's a "." on the data sample, so we should eliminate the punctuation

# empty list to store words : 
words_no_punc = []

# removing punctuation marks : 
for w in words : 
    if w.isalpha():
        words_no_punc.append(w.lower())

# print the words without punctution marks : 
print(words_no_punc)

print("\n")

# length 
print(len(words_no_punc))
```

```python
# str.isalpha() : Return True if the string is an alphabetic string, False otherwise
print("aaa".isalpha())
print("a".isalpha())
print("aaa2aaa".isalpha())
print("aaa aaa".isalpha())
```

```python
# freq distribution  
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

```python
# plot the most common
fdist.plot(10)
```

```python
# empty list to store clean words : 
clean_words = []

for w in words_no_punc : 
    if w not in stopwords : 
        clean_words.append(w)

print(clean_words)
print("\n")
print(len(clean_words))
```

```python
# frequency distribution : 
fdist = FreqDist(clean_words)

fdist.most_common(10)
```

```python
# stemming example : 

# import stemming library : 
from nltk.stem import PorterStemmer

porter = PorterStemmer()

# word-list for stemming : 
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list :
    print(porter.stem(w))
```

```python
#PoS tagging example :
sentence = "A very beautiful young lady is walking on the beach"

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

for words in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)
    
tagged_words
```

```python
# PoS (part of speech) tagging 
# PoS tagging itu dipakai supaya mesin tahu: Peran kata ini apa di kalimat?
# contoh : 
# I book a ticket (PoS Book : verb)
# I read a book   (PoS Book : Noun)

```

```python
# Extracting Noun Phrase from text 
# ? - optional character
# * - 0 or more repetations
grammar = "NP : {<DT>?<JJ>*<NN>}"
import matplotlib.pyplot as plt

# creating parser : (pattern matching)
parser = nltk.RegexpParser(grammar)

# parsing text : 
output = parser.parse(tagged_words)
print(output)

# visualize
# output.draw()
```

```python
# kata benda, kenapa penting ? karena sering menjadi : 
# 1. Subject / object
# 2. Entitas penting
# 3. Keyword untuk info extraction
# 4. Fitur untuk ML / text mining

# grammar = "NP : {<DT>?<JJ>*<NN>}"
# parser = nltk.RegexpParser(grammar)
# ini adalah pattern matching

# punctuation : tanda baca
```

```python
#Chinking example :
# * - 0 or more repetations
# + - 1 or more repetations

#Here we are taking the whole string and then
#excluding adjectives from that chunk.

grammar = r""" NP: {<.*>+} 
               }<JJ>+{"""

#Creating parser :
parser = nltk.RegexpParser(grammar)

#parsing string :
output = parser.parse(tagged_words)
print(output)

#To visualize :
#output.draw()
```

```python
# pada code di atas, digunakan untuk membuang adjective(JJ) dari noun phrase
# pada grammar 1 (2 code block atas) digunakan untuk pattern matching
# pada grammer 2 (1 code block atas) digunakan untuk mengumpulkan NP, lalu membuang yang tidak perlu (contoh disitu membuang JJ)
```

```python
# POS Tagging (Part-of-Speech Tagging) dalam Natural Language Processing (NLP) adalah proses memberi label kelas kata pada setiap kata dalam sebuah kalimat.

# Contoh kelas kata:
# - Noun (kata benda)
# - Verb (kata kerja)
# - Adjective (kata sifat)
# - Adverb (kata keterangan)
# - Pronoun
# - Preposition
# - Conjunction
# - Determiner
```

```python
#Sentence for NER :
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

#PoS tagging :
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

#print (tagged_words)

#Named Entity Recognition :
N_E_R = nltk.ne_chunk(tagged_words,binary=False)
print(N_E_R)

#To visualize :
#N_E_R.draw()
```

```python
#Sentence for NER :
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

#PoS tagging :
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

#print (tagged_words)

#Named Entity Recognition :
N_E_R = nltk.ne_chunk(tagged_words,binary=True)

print(N_E_R)

#To visualize :
#N_E_R.draw()
```

```python
#Finding antonyms :

#Empty list to store antonyms :
antonyms = []

for words in wordnet.synsets('Natural'):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
            
#Print antonyms :            
antonyms
```

```python
#Finding synonyms and antonyms :

#Empty lists to store synonyms/antonynms : 
synonyms = []
antonyms = []

for words in wordnet.synsets('New'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
            
#Print lists :
print(synonyms)
print("\n")
print(antonyms)
```

```python
#Similarity in words :
word1 = wordnet.synsets("ship","n")[0]

word2 = wordnet.synsets("boat","n")[0] 

#Check similarity :
print(word1.wup_similarity(word2))
```

```python
#Import required libraries :
from sklearn.feature_extraction.text import CountVectorizer

#Text for analysis :
sentences = ["Jim and Pam travelled by the bus:",
             "The train was late",
             "The flight was full.Travelling by flight is expensive"]
#Create an object :
cv = CountVectorizer()

#Generating output for Bag of Words :
B_O_W = cv.fit_transform(sentences).toarray()


#Total words with their index in model :
print(cv.vocabulary_)
print("\n")

#Features :
print(cv.get_feature_names_out())
print("\n")

#Show the output :
print(B_O_W)
```
