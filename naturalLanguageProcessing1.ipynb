```python
text_file = open("dataSource/nlp1source.txt")
text = text_file.read()

# print data type of the file 
print(type(text))
print("\n")
# print the text 
print(text)
print("\n")
# length of the text : 
print("Length : ", len(text))
```

```python
import nltk 
from nltk import sent_tokenize 
from nltk import word_tokenize

nltk.download("popular", quiet=True)
```

```python
nltk.download('punkt_tab')
```

```python
# Tokenize the text by sentences 
sentences = sent_tokenize(text)
print(len(sentences))

# print the sentences
sentences
```

```python
# Tokenize the text with words
words = word_tokenize(text)

# how many words are there 
print(len(words))
print("\n")

# print words :
print(words)
```

```python
# import required libraries
from nltk.probability import FreqDist 

# find the frequency : 
fdist = FreqDist(words)

fdist.most_common(10)
```

```python
# Tokenize itu bukan encoding
# yaitu adalah memecah teks menjadi unit kecil (token)
# kalimat : sentence token
# kata : word token
# contoh : text = "I love NLP."
# menjadi : ['I', 'love', 'NLP', '.']
# lalu kenapa tidak hanya sekedar .split(" ") ?
# karena jika case nya "Dr. Smith lives in the U.S."
# maka ketika di pecah menjadi ['Dr', 'Smith lives in the U', 'S'] (berdasarkan titik)
# isi nltk itu adalah kamus kata2 tersebut
```

```python
# plot the graph for fdist  : 
import matplotlib.pyplot as plt

fdist.plot(10)
```

```python
# there's a "." on the data sample, so we should eliminate the punctuation

# empty list to store words : 
words_no_punc = []

# removing punctuation marks : 
for w in words : 
    if w.isalpha():
        words_no_punc.append(w.lower())

# print the words without punctution marks : 
print(words_no_punc)

print("\n")

# length 
print(len(words_no_punc))
```

```python
# str.isalpha() : Return True if the string is an alphabetic string, False otherwise
print("aaa".isalpha())
print("a".isalpha())
print("aaa2aaa".isalpha())
print("aaa aaa".isalpha())
```

```python
# freq distribution  
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

```python
# plot the most common
fdist.plot(10)
```

```python
from nltk.corpus import stopwords

# list of stop words
stopwords = stopwords.words("english")
print(stopwords)
```

```python
# empty list to store clean words : 
clean_words = []

for w in words_no_punc : 
    if w not in stopwords : 
        clean_words.append(w)

print(clean_words)
print("\n")
print(len(clean_words))
```

```python
# frequency distribution : 
fdist = FreqDist(clean_words)

fdist.most_common(10)
```

```python
# plot the most common words on graph 
fdist.plot(10)
```

```python
# stemming example : 

# import stemming library : 
from nltk.stem import PorterStemmer

porter = PorterStemmer()

# word-list for stemming : 
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list :
    print(porter.stem(w))
```

```python
# tujuan stemming adalah membentuk kata2 menjadi kata dasar, agar di anggap sama oleh sistem 
# ada metode lain yaitu Lemmatization
# stemming lebih ke rule based (lebih cepat namun bisa keliru)
# Lemmatization berdasarkan kamus/grammar (lebih lambat namun akurat)
```

```python
# contoh Lemmatization (dari kamus)
from nltk import WordNetLemmatizer

lemma = WordNetLemmatizer()
word_list = ["am","is","are","was","were"]

for w in word_list:
    print(lemma.lemmatize(w ,pos="v"))
```

```python
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()
 
print(lemmatizer.lemmatize('studies'))
```

```python
# Bagaimana hasil lemma berubah tergantung part of speech (POS)
from nltk.stem import WordNetLemmatizer 

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('studying', pos="v"))
print(lemmatizer.lemmatize('studying', pos="n"))
print(lemmatizer.lemmatize('studying', pos="a"))
print(lemmatizer.lemmatize('studying', pos="r"))
```

```python
#PoS tagging example :
sentence = "A very beautiful young lady is walking on the beach"

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

for words in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)
    
tagged_words
```

```python
# PoS (part of speech) tagging 
# PoS tagging itu dipakai supaya mesin tahu: Peran kata ini apa di kalimat?
# contoh : 
# I book a ticket (PoS Book : verb)
# I read a book   (PoS Book : Noun)

```

```python
# Extracting Noun Phrase from text 
# ? - optional character
# * - 0 or more repetations
grammar = "NP : {<DT>?<JJ>*<NN>}"
import matplotlib.pyplot as plt

# creating parser : (pattern matching)
parser = nltk.RegexpParser(grammar)

# parsing text : 
output = parser.parse(tagged_words)
print(output)

# visualize
# output.draw()
```
