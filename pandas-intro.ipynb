```python
data = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
data
```

```python
data['a']
```

```python
population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'Florida': 19552860,
                   'New York': 19453561,
                   'Pennsylvania': 12773801}
population = pd.Series(population_dict)
population
```

```python
population['California']
```

```python
population['California':'Florida']
```
