### LDA simple process on python
So I asked ChatGPT to generate Python code to see the process of how LDA generates topics from documents.

```python
import random
from collections import defaultdict
```

```python
docs = [
["football","match","goal"],
["football","team","player"],
["election","vote","government"]
]

K = 2
```

```python
# random topic assignment
topics = []
for doc in docs:
    # each word on documents
    topics.append([random.randint(0,K-1) for _ in doc])

word_topic = defaultdict(lambda:[0]*K)
doc_topic = [ [0]*K for _ in docs]
```

```python
# print topics (should be random)
print(topics)
```

```python
# initialize count
for d,doc in enumerate(docs):
    for i,word in enumerate(doc):
        t = topics[d][i]
        word_topic[word][t]+=1
        doc_topic[d][t]+=1
```

```python
for d,doc in enumerate(docs):
    print(d,doc)
    for i,word in enumerate(doc):
        print(i,word)
    print("==============")
```

```python
for iteration in range(500):

    for d,doc in enumerate(docs):
        for i,word in enumerate(doc):

            old_topic = topics[d][i]

            # remove count
            word_topic[word][old_topic]-=1
            doc_topic[d][old_topic]-=1

            # compute probability
            probs=[]
            for k in range(K):
                p_word = word_topic[word][k] + 1
                p_doc = doc_topic[d][k] + 1
                probs.append(p_word * p_doc)

            new_topic = probs.index(max(probs))

            topics[d][i]=new_topic

            word_topic[word][new_topic]+=1
            doc_topic[d][new_topic]+=1
```
